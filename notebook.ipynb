```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

**Filtering the DataFrame**

In the segment, we filter the data to obtain specific information. Here we are looking to movies that were released between 1990 and 1999.

```python
movies = netflix_df[netflix_df['release_year'] >= 1990]
movies2 = movies[movies['release_year'] <= 1999]
movies_final = movies2[movies2['type'] == 'Movie']

print(movies_final)
```

Next we use the 'value_counts()' method to find the frequencies of distinct values from 'movies_final'.

```python
duration_count = movies_final['duration'].value_counts()
print(duration_count)

duration = 94
```

As we can see in the Series above, the most frequent duration is 94 mins for 7 movies.

**Counting the number fo Short Action Movies**

In the following cell, we further filter out our data by genre and count the number of movies with duration less than 90 mins.

```python
short_movies = movies2[movies2['duration'] < 90]
short_movies2 = short_movies[short_movies['genre'] == 'Action']
short_movie_count = len(short_movies2)
print(short_movie_count)
```
